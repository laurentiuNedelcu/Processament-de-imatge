#### Libraries we are going to use for this notebook

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import ndimage, signal
from skimage import io, filters, color, exposure
```

#### Functions that we are going to use for this notebook

```python
"""
Function to roll the respective x and y distance in the image.

@params:
    img: image in performing the roll
    dispX: distance to move on the coordinate x
    dispY: distance to move on the coordinate y
"""
def roll(img, dispX, dispY):
    
    copyX = img[:, :-dispX]
    img = np.concatenate((img[:,-dispX:img.shape[1]],copyX),axis=1)
    
    copyY = img[:-dispY, :]
    img2 = np.concatenate((img[-dispY:img.shape[0],:],copyY),axis=0)
    
    return img2

"""
Function to calculate the correlation position and the respective distances that the images are moved.

@params:
    NCC: correlation between both images.
"""

def calculate_coord(NCC):
    
    # Get max. correlation position
    y_coord, x_coord = np.unravel_index(np.argmax(NCC), NCC.shape)
    
    # Get center of original patch
    y_org_coord, x_org_coord = (NCC.shape[0]//2, NCC.shape[1]//2)
    
    # Calculate distance between correlation coords and center
    dist_x , dist_y = (x_org_coord - x_coord, y_org_coord - y_coord)
    
    return (x_coord, y_coord, x_org_coord, y_org_coord, dist_x, dist_y)
    

"""
Function to disploy the resulting images with their respective title.
@params:
    img_1: first image patched
    img_2: second image patched
    x1: x coordinate of the original image.
    y1: y coordinate of the original image.
    x2: x coordinate of the correlation position.
    y2: y coordinate of the correlation position.
    dx: distance between x1 and x2.
    dy: distance between y1 and y2.
    NCC: correlation between both images.
    size: size of the image. Default = 15.
    title: title of the respective channel.
"""
def plot_imgs(img_1, img_2, x1, y1, x2, y2, dx, dy, NCC, size=15, title=""):
    
    plt.figure(figsize=(size,size))
    plt.subplot(131)
    plt.title("Original image (" + str(x1) + "," + str(y1) + ")")
    plt.imshow(img_1, cmap='jet')
    plt.subplot(132)
    plt.title("Displaced image " + title + " (" + str(x2) + "," + str(y2) + ")")
    plt.imshow(img_2, cmap='jet')
    plt.subplot(133)
    plt.title("Correlation between both images (" + str(dx) + "," + str(dy) + ")")
    plt.imshow(NCC, cmap='jet')
    plt.show()
    

"""
Function that calculates the cross correlation between 2 channels of an image.

@params:
    image_1: image with the first channel.
    image_2: image with the second channel.
    x: x coordinate on which the cross correlation should be centered
    y: y coordinate on which the cross correlation should be centered
    size: block size (px). Default = 15
    step: size of the search area. Default = 3
    title: title to display on the plot. Default = ""
"""
def cross_correl(image_1, image_2, x, y, size=15, step=3, title="", display=True):
    # Generate image patches
    image_1_patch = image_1[y-size:y+size+1, x-size:x+size+1]
    image_2_patch = image_2[y-size-step:y+size+step+1, x-size-step:x+size+step+1]
    
    # Normalize paches
    image_1_patch = (image_1_patch - image_1_patch.min()) / image_1_patch.max()
    image_1_patch -= image_1_patch.mean()
    
    image_2_patch = (image_2_patch - image_2_patch.min()) / image_2_patch.max()
    image_2_patch -= image_2_patch.mean()   
    
    # Apply correlation between patches
    NCC = signal.correlate2d(image_2_patch.astype('float'), image_1_patch.astype('float'),
                             mode='full', boundary='fill', fillvalue=0)
    
    # Get the coordinates and desplacements
    x_coord, y_coord, x_org_coord, y_org_coord, dist_x, dist_y = calculate_coord(NCC)
    
    #Display
    if display:
        plot_imgs(image_1_patch, image_2_patch, x_org_coord, y_org_coord, x_coord, y_coord, dist_x, dist_y,
              NCC, title=title)
    
    # Generate corrected image
    return roll(image_2, dist_x, dist_y)



def cross_corr_pyramidal(image_1, image_2, x, y, size, step, n, size_big, step_big, title="", display=True):
    
    # Generate image patches
    y1_str = y - size - step
    y1_end = y + size + step + 1
    x1_str = x - size - step
    x1_end = x + size + step + 1
    
    image_1_patch = image_1[y1_str:y1_end, x1_str:x1_end]
    
    y2_str = y - size_big - step_big
    y2_end = y + size_big + step_big + 1
    x2_str = x - size_big - step_big
    x2_end = x + size_big + step_big + 1
    
    image_2_patch = image_2[y2_str:y2_end, x2_str:x2_end]
    
    # Normalize patches   
    image_1_patch = (image_1_patch - image_1_patch.min()) / image_1_patch.max()
    image_1_patch -= image_1_patch.mean()
    
    # Normalize patches
    image_2_patch = (image_2_patch - image_2_patch.min()) / image_2_patch.max()
    image_2_patch -= image_2_patch.mean()
    
    # Apply gaussian filter
    image_1_filtered = filters.gaussian(image_1_patch, 1)
    image_2_filtered = filters.gaussian(image_2_patch, 1)
    
    # Apply sampling
    image_1_sample = image_1_filtered[::n,::n]
    image_2_sample = image_2_filtered[::n,::n]
    
    # Apply correlation between patches
    NCC_sample = signal.correlate2d(image_2_sample.astype('float'), image_1_sample.astype('float'),
                                    mode='full', boundary='fill', fillvalue=0)
    
    # Get the coordinates and desplacements
    x_coord, y_coord, x_org_coord, y_org_coord, dist_x, dist_y = calculate_coord(NCC_sample)
    
    #Display
    if display:
        plot_imgs(image_1_patch, image_2_patch, x_org_coord, y_org_coord, x_coord, y_coord, dist_x, dist_y,
              NCC_sample, title="title")
    
    # Generate corrected image aproximation
    image = roll(image_2, dist_x * n, dist_y * n)
    
    return cross_correl(image_1, image, x, y, size=size, step=step, title=title,display=display)
```

## Week 1: R, G, B channel spacing and alignment
#### 1.1 Reconstruction

```python
# Load image
img = io.imread("images/image.png")

# Resize the image for speed-up
print('The original image has size ', img.shape)
img_np = img[::10,::10]
print('The resized image has a final size of ', img_np.shape)

# Defining the position of the target object
x1 = 20; y1 = 20
x2 = 25; y2 = 333
x3 = 24; y3 = 647
w = 336; h = 302
im1 = img_np[y1-2:y1+h-2, x1-1:x1+w-1]
im2 = img_np[y2-2:y2+h-2, x2-1:x2+w-1]
im3 = img_np[y3-2:y3+h-2, x3-1:x3+w-1]
I1 = 255 * im1.astype('double') / im1.max()
I2 = 255 * im2.astype('double') / im2.max()
I3 = 255 * im3.astype('double') / im3.max()
RGB = np.zeros([h,w,3], dtype=type(img_np[0,0]))
RGB[:,:,0] = I3
RGB[:,:,1] = I2
RGB[:,:,2] = I1
```

#### 1.2 Alignment
We will align the rgb channels of an image using correlations. In this way, we can join all the channels that have been moved.  

The process consists on obtaining the patches of two images (each one will have only one channel of the original image) and making a correlation between them. Next, we look for the maximum correlation point and finally we calculate the distance from the center point of both images (each one with its channel) to apply the corresponding offset.

```python
rolled_green = cross_correl(RGB[:,:,0], RGB[:,:,1], x=140, y=165, title= "Green Channel")
rolled_blue = cross_correl(RGB[:,:,0], RGB[:,:,2], x=140, y=165, title= "Blue Channel")
```

```python
RGB_rolled = np.zeros([h,w,3], dtype=type(img_np[0,0]))
RGB_rolled[:,:,0] = RGB[:,:, 0]
RGB_rolled[:,:,1] = rolled_green
RGB_rolled[:,:,2] = rolled_blue

plt.figure(figsize = (15,15))
plt.subplot(121)
plt.title("Image before alignment",fontsize=20)
plt.imshow(RGB)
plt.subplot(122)
plt.title("Image after alignment",fontsize=20)
plt.imshow(RGB_rolled)
plt.show()
```

## Week 2: Alignment with pyramidal decomposition
In this part we will see a faster way to get the image alignment. In order to accomplish this we will implement a pyramid search. First we reconstruct the original image.

```python
# Load image
img = io.imread("images/image.png")

# Resize the image for speed-up
print('The original image has size ', img.shape)
img_np = img[::10,::10]
print('The resized image has a final size of ', img_np.shape)

# Defining the position of the target object
x1 = 20; y1 = 20
x2 = 25; y2 = 333
x3 = 24; y3 = 647
w = 336; h = 302
im1 = img_np[y1-2:y1+h-2, x1-1:x1+w-1]
im2 = img_np[y2-2:y2+h-2, x2-1:x2+w-1]
im3 = img_np[y3-2:y3+h-2, x3-1:x3+w-1]
I1 = 255 * im1.astype('double') / im1.max()
I2 = 255 * im2.astype('double') / im2.max()
I3 = 255 * im3.astype('double') / im3.max()
RGB = np.zeros([h,w,3], dtype=type(img_np[0,0]))
RGB[:,:,0] = I3
RGB[:,:,1] = I2
RGB[:,:,2] = I1
```

In order to create the pyramidal search, we will start from a subsampling of size n. From here, we will do a search on a large pattern area. In this way, then we can do a search without subsampling of a smaller area.
  
The process itself consists of obtaining the image patch and the search patch for the displaced image using a larger window. Then a Gaussian filter will be applied to the patches like Anti-aliasing. Next, a sampling will be applied to reduce the measurement and the correlation between them will be made. Once we obtain the correlation, we will find the maximum point and calculate the distance from the central point to the original one and the correlation point. Next, we apply the corresponding offset and use the resulting image to apply the same process as in the first part.

```python
%%time
x = 140
y = 165
size=15
step=3
size_big=60
step_big=3
n=2

rolled_green_sample = cross_corr_pyramidal(RGB[:,:,0], RGB[:,:,1], x, y, size, step, n, size_big, step_big)
rolled_blue_sample = cross_corr_pyramidal(RGB[:,:,0], RGB[:,:,2], x, y, size, step, n, size_big, step_big)
```

```python
RGB_sample = np.zeros([h,w,3],dtype=type(img_np[0,0]))
RGB_sample[:,:,0] = RGB[:,:,0]
RGB_sample[:,:,1] = rolled_green_sample
RGB_sample[:,:,2] = rolled_blue_sample

plt.figure(figsize = (20,20))
plt.subplot(121)
plt.title("Original image")
plt.imshow(RGB)
plt.subplot(122)
plt.title("Resulting image")
plt.imshow(RGB_sample)
plt.show()
```

## Week 3: Improve image appearance

Finally, we will improve the appearance of our resulting image. In our case we will make a balance of colors.

To achieve this we will do a white balance taking the maximum values of the image as a reference to move the histogram of the entire image. That is, we will move this reference to the value 255 that corresponds to white, in this way we will make the pixel closest to the white value really white.

```python
# Converting the image to grayscale and getting the closest white pixel
RGB_sample_grayscale = color.rgb2gray(RGB_sample)
max_white_y, max_white_x = np.unravel_index(np.argmax(RGB_sample_grayscale), RGB_sample_grayscale.shape)

RGB_white = RGB_sample.copy()

# Applying changes
RGB_white[0] = RGB_sample[0] * (255 / RGB_sample[max_white_y, max_white_x, 0])
RGB_white[1] = RGB_sample[0] * (255 / RGB_sample[max_white_y, max_white_x, 1])
RGB_white[2] = RGB_sample[0] * (255 / RGB_sample[max_white_y, max_white_x, 2])

# Display
plt.figure(figsize = (20,20))
plt.subplot(121)
plt.title("Image aligned")
plt.imshow(RGB_sample.astype(np.uint8))
plt.subplot(122)
plt.title("White image")
plt.imshow(RGB_white.astype(np.uint8))
plt.show()
```

Moreover we can use some skimage functionalities in order to change aspects of the image. For example exposure if we want to change the brightness of the image.

```python
# Display
plt.figure(figsize = (20,20))
plt.subplot(131)
plt.title("Original")
plt.imshow(RGB_white.astype(np.uint8))
plt.subplot(132)
plt.title("Brighter")
plt.imshow(exposure.rescale_intensity(RGB_white, out_range=(60,255)).astype(np.uint8))
plt.subplot(133)
plt.title("Darker")
plt.imshow(exposure.rescale_intensity(RGB_white, out_range=(0,150)).astype(np.uint8))
plt.show()
```
